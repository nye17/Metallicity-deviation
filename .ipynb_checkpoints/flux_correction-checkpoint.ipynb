```python
# ÈÄöËøáBPTÁ≠õÈÄâÔºåÂØªÊâæÂá∫pure star forming spaxels
# Ëé∑ÂèñÊØè‰∏™spaxelÁöÑÁâ©ÁêÜÂèÇÊï∞ÔºöÊï∞ÁªÑ‰∏≠‰ΩçÁΩÆ„ÄÅflux„ÄÅsurface mass density„ÄÅE(B-V)Á≠â
# ÁªìÊûúÂ∞ÜÂ≠òÂÇ®Âú®/root/.jupyter/cosmos/code_git/Metallicity-deviation/tmp_storage/galid_ij_den_NO_logOiiioii_logniioii_star_forming.dat
#######################################
#######################################
#######################################
# # Ê∏ÖÁ©∫ÂÜÖÂ≠òÊ±†ÂΩì‰∏≠ÁöÑÂ∑®Â§ßÊï∞ÁªÑ
# %xdel galid_den_NO 

import time
from copy import copy as copy
import warnings
from os.path import join

import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt    # ÁîªÂõæ[MatlabÂΩ¢Âºè]

from astropy.io import fits    # ËØªÂèñfitsÊñá‰ª∂
from marvin.tools import Maps

warnings.filterwarnings('ignore')    # ‰∏çÊòæÁ§∫warning‰ø°ÊÅØ
# ÊâìÂºÄfirefly fitsÊñá‰ª∂
vac = fits.open('/root/cosmos/data/manga_firefly-v2_4_3-STELLARPOP.fits') 
basic = copy(vac[1].data)    # galaxyÂü∫Êú¨‰ø°ÊÅØ
# ‰∏ãËΩΩÁöÑÁ¶ªÁ∫øMarvinÊï∞ÊçÆË∑ØÂæÑ
path_marvin_data = '/root/bash_file/marvin_vor10_fits'
# Â≠òÂÇ® Êï∞ÁªÑ‰ΩçÁΩÆ, Nii6584,Oiii5007,Oii3729,Oii3726,Oiii/Oii,Nii/Oii 10‰∏™Èáè
#galid_den_NO = np.zeros([basic['plate'].size*2800,12],dtype='float32')-9999
galid_den_NO = np.zeros([basic['plate'].size*2800,12])-9999
# surface mass density
sur_density_all = copy(vac['SURFACE_MASS_DENSITY_VORONOI'].data[:,:,0])

bin1d = vac[4].data    # Ëé∑Âèñbin_idÊï∞ÁªÑÔºå‰∏∫‰∫ÜÂåπÈÖçxyÂùêÊ†á‰ΩçÁΩÆ
bin_2d = vac[5].data    # Ëé∑Âèñbin_idÊï∞ÁªÑÔºå‰∏∫‰∫ÜÂåπÈÖçxyÂùêÊ†á‰ΩçÁΩÆ

manga_EBV = vac[10].data    # Ëé∑ÂèñEÔºàB-VÔºâ

# Ëé∑Âèñplate-ifuÔºå‰∏∫‰∫ÜÈÖçÂêàMarvinÁöÑÂåÖ
plate_np = basic['plate']
ifu_np = basic['ifudsgn']
plate_np = plate_np.astype(np.int64)    # Êï∞ÊçÆËΩ¨Âåñ‰∏∫Êï¥Âûã
#ifu_np = ifu_np.astype(np.int64)    # Êï∞ÊçÆËΩ¨Âåñ‰∏∫Êï¥Âûã

lalala_start = time.time()    # ÂºÄÂßãËÆ°Êó∂
# for i in range(0,basic['plate'].size):
i = 0    # ÊòüÁ≥ªÁöÑÊï∞ÁªÑ‰ΩçÁΩÆ: i
j = 0    # 4675*2800ÁöÑÊï∞ÁªÑ‰ΩçÁΩÆ: j
#for j in range(0,1):    # 4675*2800ÁöÑÊï∞ÁªÑ‰ΩçÁΩÆ: j
#while j<4675*2800:
#while j<10:
#while j<basic['plate'].size*sur_density_all[0].size:
# lala=0
# ÊéßÂà∂Âæ™ÁéØÊúÄÂ§ßÂÄº
# j_end = basic['plate'].size*sur_density_all[0].size
j_end = 17*sur_density_all[0].size
while j<j_end:
    start_for = time.time()    #Ëé∑ÂèñÊ≠§Ê¨°Âæ™ÁéØÂºÄÂßãÊó∂Èó¥
    
    
    
    # ÊØè‰∏™ÊòüÁ≥ªÂèñÂÄº
    plate = plate_np[i]    # Ëé∑Âèñplate
    ifu = ifu_np[i]    # Ëé∑Âèñifu
    galid = (basic['plate']==plate)&(basic['ifudsgn']==str(ifu))    # ÊòüÁ≥ª‰ΩçÁΩÆÊï∞ÁªÑÔºå‰ΩøÁî®plate&ifuËøõË°åÂÆö‰ΩçÔºåÂæóÂà∞True FalseÊï∞ÁªÑ
    bin_1d = bin1d[galid,:,0][0]    # Ëé∑Âèñ‰∏ÄÁª¥ÁöÑbinÊï∞ÊçÆ
    #sur_density = sur_density_all[galid][0]
    
    filename = join(path_marvin_data, 'manga-{}-MAPS-VOR10-GAU-MILESHC.fits.gz'.format(''.join([str(plate),'-',str(ifu)])))
    new_maps = Maps(filename=filename)
    ha = new_maps.emline_gflux_ha_6564    # Ëé∑ÂèñHŒ±6564
    hb = new_maps.emline_gflux_hb_4862    # Ëé∑ÂèñHŒ≤4862
    nii = new_maps.emline_gflux_nii_6585    # Ëé∑ÂèñNii_6584
    oii7 = new_maps.emline_gflux_oii_3727    # Ëé∑ÂèñOii_3726
    oii9 = new_maps.emline_gflux_oii_3729    # Ëé∑ÂèñOii_3729
    oiii = new_maps.emline_gflux_oiii_5008    # Ëé∑ÂèñOiii_5007
    
    #lg_oiii_oii = np.log10(oiii/(oii7+oii9))
    #lg_nii_oii = np.log10(nii/(oii7+oii9))
    
    
    
    # ËÆ°ÁÆóÈáëÂ±û‰∏∞Â∫¶
    RR = (oii7+oii9) / nii    # Oii/Nii
    RR2 = (oii7+oii9) / oiii    # Oii/Oiii
    
    # ËÆ°ÁÆómask
    masks_bpt, __, __ = new_maps.get_bpt(show_plot=False)    # ‰ΩøÁî®bptÂõæÁ≠õÈÄâstar-formingÂå∫Âüü
    mask_non_sf = ~masks_bpt['sf']['global'] * RR.pixmask.labels_to_value('DONOTUSE') * RR2.pixmask.labels_to_value('DONOTUSE')
    mask_bad_data = RR.pixmask.get_mask(['NOCOV', 'UNRELIABLE', 'DONOTUSE'])
    mask_bad_data2 = RR2.pixmask.get_mask(['NOCOV', 'UNRELIABLE', 'DONOTUSE'])
    min_snr = 10.    #‰ø°Âô™ÊØîËÆæÂÆö‰∏∫10
    mask_ha_low_snr = (np.abs(ha.value * np.sqrt(ha.ivar)) < min_snr)
    mask_hb_low_snr = (np.abs(hb.value * np.sqrt(hb.ivar)) < min_snr)
    mask_nii_low_snr = (np.abs(nii.value * np.sqrt(nii.ivar)) < min_snr)
    mask_oii7_low_snr = (np.abs(oii7.value * np.sqrt(oii7.ivar)) < min_snr)
    mask_oii9_low_snr = (np.abs(oii9.value * np.sqrt(oii9.ivar)) < min_snr)
    mask_oiii_low_snr = (np.abs(oiii.value * np.sqrt(oiii.ivar)) < min_snr)
    
    new_mask = mask_non_sf | mask_bad_data | mask_bad_data2 | mask_ha_low_snr | mask_hb_low_snr | mask_nii_low_snr | mask_oii7_low_snr | mask_oii9_low_snr| mask_oiii_low_snr     # ÂæóÂà∞ÊúÄÂêéÁöÑmask
    #print("BPTÂàÜÁ±ªÂæóÂà∞ÁöÑmaskÁü©ÈòµÂ§ßÂ∞è‰∏∫",new_mask.shape)
    # star forming spaxel ‰∏™Êï∞
    #num_sfr = np.argwhere(lala1_sfr==0).size/2
    
    
    # ÊØè‰∏™spaxelÂèñÂÄº
    image_sz = np.shape(nii)[0]    # ÂõæÂÉèËæπÈïø
    # ‰∏¥Êó∂ÊµãËØïÔºåÁîªÂõæÊï∞ÁªÑ
    #print("ÂéüÂõæÂÉèÂ§ßÂ∞è‰∏∫Ôºö ",image_sz,'*',image_sz)
    maps = np.zeros((image_sz,image_sz))-9999    # ÂàùÂßãÂåñÁîªÂõæÊï∞ÁªÑ
    for y in range(0,image_sz):
        for x in range(0,image_sz):
            id_bin = (bin_1d==bin_2d[galid,y,x])    # ÊêúÁ¥¢‰∫åÁª¥binÁöÑÂÄºÂú®‰∏ÄÁª¥binÊï∞ÁªÑ‰∏≠ÁöÑ‰ΩçÁΩÆ
            #if (len(bin_1d[id_bin])==1):
            if (len(bin_1d[id_bin])==1)&(new_mask[y,x]==0):    # ÊåëÈÄâÂá∫ ‰∏ÄÁª¥2800ÁöÑbinÂíå‰∫åÁª¥76*76ÁöÑbinÁõ∏ÂêåÁöÑÊï∞ÂÄºÔºå‰ª•ÂèäsfÁöÑspaxel
                id_bin_num = np.where(id_bin==True)[0][0]    # 2800ÈïøÂ∫¶Êï∞ÁªÑ‰∏≠ÁöÑ‰ΩçÁΩÆ
                #print( np.where(id_bin==True))
                # spaxelÊâÄÂú®ÁöÑÊòüÁ≥ªÁöÑÊï∞ÁªÑ‰ΩçÁΩÆ
                galid_den_NO[j,0] = i
                # spaxelÂú®2800Êï∞ÁªÑ‰∏≠ÁöÑ‰ΩçÁΩÆ
                galid_den_NO[j,1] = id_bin_num
                # ÊòüÁ≥ª‰∏≠ÁöÑy‰ΩçÁΩÆ
                galid_den_NO[j,2] = y
                # ÊòüÁ≥ª‰∏≠ÁöÑx‰ΩçÁΩÆ
                galid_den_NO[j,3] = x
                # spaxel's surface mass density
                #galid_den_NO[j,4] = sur_density[id_bin]
                galid_den_NO[j,4] = denty_yes_corret[y,x]
                # ha
                galid_den_NO[j,5] = ha.value[y,x]
                # hb
                galid_den_NO[j,6] = hb.value[y,x]
                # nii 6584
                galid_den_NO[j,7] = nii.value[y,x]
                # oiii 5007
                galid_den_NO[j,8] = oiii.value[y,x]
                # oii 3729
                galid_den_NO[j,9] = oii9.value[y,x]
                # oii 3726
                galid_den_NO[j,10] = oii7.value[y,x]
                # ÂèñÂá∫E(B-V)
                galid_den_NO[j,11] = manga_EBV[i,id_bin_num]
                
                j+=1
                
                #maps[y,x] = oii7.value[y,x]    # ‰∏¥Êó∂ÊµãËØïÔºåÁîªÂõæÊï∞ÁªÑ
                maps[y,x] = denty_yes_corret[y,x]
                
                
                
    end_for = time.time()    # Ëé∑ÂèñÊ≠§Ê¨°Âæ™ÁéØÁªìÊùüÊó∂Èó¥
    print('Áî®Êó∂: ',end_for-start_for,"  Â∑≤ÁªèËÆ°ÁÆóÂà∞Á¨¨ ",i+1," ‰∏™ÊòüÁ≥ª   ","Ââ©‰Ωô ",4675-i-1,' ‰∏™ÊòüÁ≥ª')
    
    # Ë∑≥ËøáÊó†ÊïàÊï∞ÊçÆÔºåÁõ¥Êé•ËøõÂÖ•‰∏ã‰∏Ä‰∏™ÊòüÁ≥ª
    i += 1
    j = i*2800-1
    
    # ‰∏¥Êó∂ÊµãËØï
    #print("j = ",j)
    #print("j‰∏é‰∏ä‰∏ÄÊ¨°ÁöÑÂ∑ÆÂÄº = ",j-lala)
    #lala =j
    
    # ÈôêÂà∂Âæ™ÁéØÊ¨°Êï∞ÔºåÈò≤Ê≠¢Ë∂ÖËøáÊï∞ÁªÑÂÆûÈôÖÂ§ßÂ∞è
    if j>=j_end-1:
        break
    p3d_data.close()
    
    
lalala_end = time.time()
print('\n\n\nÊÄªÁî®Êó∂Ôºö ',str(lalala_end-lalala_start))

# Â∞ÜÂæóÂà∞ÁöÑÁªìÊûúËøõË°å‰øùÂ≠ò (4675*2800,10)
# Êó†ÊïàÊï∞ÊçÆ‰∏∫-9999
# np.savetxt("/root/.jupyter/cosmos/code_git/Metallicity-deviation/tmp_storage/MaNGA_star_forming_spaxels_data.dat", galid_den_NO)               

# ÁÆÄÂçïÁîªÂõæÊµãËØï
fig, ax = plt.subplots()
p = ax.imshow(maps , origin='lower',vmin=0,vmax=1)
plt.show()


vac.close() 
```

### ÂæóÂà∞Ê≠£Á°ÆÁöÑ Á≠õÈÄâÂá∫ÁöÑÊòüÁ≥ªÁöÑ spaxel ÁöÑ surface mass density
1. > Â∞Ü surface mass density ÁöÑÂçï‰Ωç $M_{\odot}*spaxel^{-2}$ ËΩ¨Êç¢‰∏∫ $M_{\odot}*kpc^{-2}$   
>>$log_{10}(M_{\odot}*spaxel^{-2}) - (2*log_{10}(distance_{com})+2*log_{10}(\pi/1296))=log_{10}(M_{\odot}*kpc^{-2})$ 
   
1. > ËÄÉËôëÊ§≠ÂúÜÁéáÔºåÂÅöÈÄÜÊò†Â∞Ñ   
>>$density_{obs}*b/a = density_{in}$ 

```python
import numpy as np
from math import *
from astropy.cosmology import LambdaCDM
# Âà©Áî®Á∫¢ÁßªÂíåspaxelÂ§ßÂ∞èËÆ°ÁÆóÊØè‰∏™spaxelÁöÑÂÆûÈôÖÂ§ßÂ∞è„ÄêÂçï‰Ωç‰∏∫kpc^2„Äë
# ËÄÉËôëÊ§≠ÂúÜÁéá b/aÔºåÂ∞Üsurface mass density ÂÅöÈÄÜÊò†Â∞ÑÔºåÂç≥ new_density = old_density*b/a

# ÂÆáÂÆôÂ≠¶Ê®°Âûã
cosmo = LambdaCDM(H0=70., Om0=0.3, Ode0=0.7)

# ËΩΩÂÖ•Êï∞ÊçÆ
# spaxel_data = np.loadtxt("./tmp_storage/galid_ij_den_NO_logOiiioii_logniioii_star_forming_v2.0.dat")
# ËØªÂèñÂÖ®ÈÉ®ÁöÑstar-forming spaxelÊï∞ÊçÆ
spaxel_data = np.loadtxt("./tmp_storage/MaNGA_star_forming_spaxels_data_v2.0.dat")
# ËØªÂèñ PIPE3D surface mass density
spaxel_density = np.loadtxt("./tmp_storage/MaNGA_PIPE3D_spaxels_density.dat")

# ÂéªÈô§Êó†ÊïàÊï∞ÊçÆ
spaxel_density = spaxel_density[np.where(spaxel_data[:,0]!=-9999)]
spaxel_data = spaxel_data[np.where(spaxel_data[:,0]!=-9999)]
id_manga = spaxel_data[:,0].copy()    # MaNGA spaxelÁöÑÊòüÁ≥ªÊï∞ÁªÑ‰ΩçÁΩÆ
size_spaxels = np.size(spaxel_data[:,0])    # spaxelsÊÄªÊï∞Èáè
# Ê†πÊçÆ‰∏äÊñπÁ≠õÈÄâÁªìÊûúÔºö1277‰∏™ÊòüÁ≥ªÔºåÊèêÂèñÂÆÉ‰ª¨ÁöÑredshift, b/a, density
b_over_a = b_over_a[sel_galaxy[0]]
z = z[sel_galaxy[0]]

# Âà©Áî® Á∫¢Áßª ËøõË°åcomoving distance ËÆ°ÁÆó
com_dis = cosmo.comoving_distance(z).value
# ÂéüÊù•ÁöÑdensityÊï∞ÊçÆÁöÑÂçï‰ΩçËΩ¨Êç¢‰∏∫ÁõÆÊ†áÂçï‰ΩçÔºåÈúÄË¶ÅÂä†ÁöÑÂÄº
add_orig_density = 2*np.log10(com_dis) + 2*np.log10(math.pi/1296)

sel_manga = manga_match_nsa[sel_galaxy[0],0]
sel_nsa = manga_match_nsa[sel_galaxy[0],1]
```

```python
#@Junde Chen 2019 jundechen@sjtu.edu.cn
#extinction/reddening correction routine

import numpy as np

def _isiterable(obj):
    """
    check if object is iterable
    """
    try:
        iter(obj)
        return True
    except TypeError:
        return False
#################################################################################
def calzetti00(wave,Av,Rv):
    """
    Calzetti et al (2000) extinction function law.

    Parameters:
    ===========
    wave : wavelength in Angstroms
    Av : v-band extinction
    Rv : extinction coefficient, Av/E(B-V), 4.05 for starburst galaxies
    
    Returns:
    ========
    extinctions at each input wavelength
    
    """
    x = wave/10**4 #convert Angstrom to um
    if _isiterable(x):
        return np.array([_calzetti00_module(i,Rv) for i in x])*Av/Rv
    else:
        return _calzetti00_module(x,Rv)*Av/Rv

# k'(Œª)ÂÆö‰πâÔºåÂàÜÊÆµÂáΩÊï∞Ôºå‰ª• 0.63‰∏∫ÂàÜÁïåÁÇπ
def _calzetti00_module(x,Rv):
    if 0.63<=x:
        return 2.659*(-1.857+1.040/x)+Rv
    elif x<0.63:
        return 2.659*(-2.156+1.509/x-0.198/x/x + 0.011/x/x/x)+Rv
    
#################################################################################
def chen_red_corr_calzetti00(wave,flux,oc_ratio,c1,c2,Rv=4.05):
    """
    correct intrinsic reddening using Calzetti00 extinction curve

    Parameters:
    ===========
    wave :  input wavelength in Angstrom
    flux :  the observed flux at input wave
    oc_ratio=r_cri/r_obs : r_obs: flux ratio of line1 and line2, r_cri: intrinsic value of line1/line2
    c1,c2 : line center of 1 and c2
    Rv : extinction coefficient, Av/E(B-V), 4.05 for starburst galaxies

    Returns:
    ===========
    corrected flux
    """
    x = wave
    k1 = calzetti00(c1,1.0,Rv)
    k2 = calzetti00(c2,1.0,Rv)
    kx = calzetti00(x ,1.0,Rv)
    A_lamda = 2.5*kx/(k1-k2)*np.log10(oc_ratio)
    return flux*10**(0.4*A_lamda)
def red_corr_calzetti00(wave,flux,oc_ratio,c1,c2,Rv=4.05):
    """
    correct intrinsic reddening using Calzetti00 extinction curve

    Parameters:
    ===========
    wave :  input wavelength in Angstrom
    flux :  the observed flux at input wave
    oc_ratio=r_cri/r_obs : r_obs: flux ratio of line1 and line2, r_cri: intrinsic value of line1/line2
    c1,c2 : line center of 1 and c2
    Rv : extinction coefficient, Av/E(B-V), 4.05 for starburst galaxies

    Returns:
    ===========
    corrected flux
    """
    x = wave
    k1 = _calzetti00_module(c1,Rv)
    k2 = _calzetti00_module(c2,Rv)
    kx = _calzetti00_module(x ,Rv)
    A_lamda = 2.5*kx/(k1-k2)*np.log10(oc_ratio)
    return flux*10**(0.4*A_lamda)

def old_bm_EBV(lam_ha,lam_hb,flux_ha,flux_hb,A_v,R_v):
    k_ha = calzetti00(lam_ha,A_v,R_v)
    k_hb = calzetti00(lam_hb,A_v,R_v)
    flux_ha_o_hb = flux_ha/flux_hb
    A_ha_hb = 2.5*np.log10(2.86/flux_ha_o_hb)
    balmer_EBV = A_ha_hb/(k_ha-k_hb)
    return balmer_EBV
    
def bm_EBV(lam_ha,lam_hb,flux_ha,flux_hb,R_v):
    k_ha = _calzetti00_module(lam_ha,R_v)
    k_hb = _calzetti00_module(lam_hb,R_v)
    flux_ha_o_hb = flux_ha/flux_hb
    balmer_EBV = 1/(k_ha-k_hb)*2.5*np.log10(2.86/flux_ha_o_hb)
    return balmer_EBV
```

```python
"""
Balmer decrement ha/hb=2.86
ÂÆûÁé∞ intrinsic extinction correction

ÂáΩÊï∞ËØ¥ÊòéÔºö
calzetti00_dered(wave,flux,ha_lam,hb_lam,ha_flux,hb_flux,Rv,unit):
    dereddeningÂáΩÊï∞ÔºåËøõË°åextinction correction

calzetti00k_wave(wave,Rv,unit)Ôºö
    ËÆ°ÁÆó k(lambda)

calzetti00_EBV(ha_lam,hb_lam,ha_flux,hb_flux,Rv,unit):
    ÂèØ‰ª•ÂæóÂà∞E(B-V)

calzetti00_extinction(wave,A_v,Rv,unit):
    Ëøô‰∏™ÊòØ A(lambda)ÔºåÁî®Êù•Áîªextinction curve
    Âç≥‰ª§A_v=1ÔºåÂàôÂæóÂà∞Ê∂àÂÖâÊõ≤Á∫ø A(lambda)/A_v

ÂèÇÊï∞ËØ¥ÊòéÔºö
waveÔºöÊ≥¢Èïø„ÄêÂèØ‰ª•Âçï‰∏™Êï∞Â≠óÔºå‰πüÂèØ‰ª•‰∏Ä‰∏™Êï∞ÁªÑ„Äë
fluxÔºöÈúÄË¶ÅËøõË°å‰øÆÊ≠£ÁöÑflux
ha_lamÔºöHŒ±Ê≥¢Èïø
hb_lamÔºöHŒ≤Ê≥¢Èïø
ha_fluxÔºöHŒ±ËßÇÊµãÂà∞ÁöÑflux
hb_fluxÔºöHŒ≤ËßÇÊµãÂà∞ÁöÑflux
RvÔºöcalzetti00ÂèñÁöÑ4.05
unit: aa[Angstrom] or um[micron]„ÄêÊ≥¢ÈïøÁöÑÂçï‰ΩçÂüÉÊàñËÄÖÂæÆÁ±≥„Äë

"""
import numpy as np

# Âà§Êñ≠ËæìÂÖ•ÁöÑfluxÊòØÊï∞ÁªÑËøòÊòØÂçï‰∏™Êï∞Â≠ó
def isiterable(obj):
    #a number or an array
    try:
        iter(obj)
        return True
    except TypeError:
        return False

# k(lambda), only a number of the x=1/lambda, not an array
# Âè™ËÉΩËæìÂÖ•Êï∞Â≠óÔºå‰∏çËÉΩÊòØÊï∞ÁªÑ
def calzetti00k_x(x,Rv):
	if x>1./0.63:
		# wavelength UV
		k = 2.659 * (((0.011*x - 0.198)*x + 1.509)*x - 2.156) + Rv
	elif 0<=x<=1./0.63:
		k = 2.659 * (1.040*x - 1.857) + Rv
	elif x<0:
		raise ValueError("wavelength negative error")
	return k

# standard form‰∏≠ÁöÑ k(lambda)Ê≠§ÂáΩÊï∞‰∏∫‰∫ÜÂêåÊó∂Êé•Êî∂Êï∞Â≠óÂíåÊï∞ÁªÑÂèòÈáè
def calzetti00k_wave(wave,Rv,unit):
	# judge the unit
	if unit == 'um':
		pass
	elif unit == 'aa':
		wave = wave/1e4
	else:
		raise ValueError("unrecognized unit \nunit = 'aa' or 'um' ")
			
	# x = 1/lambda
	x = 1./wave
	
	if isiterable(x):
		return np.array([calzetti00k_x(i,Rv) for i in x])
	else:
		return calzetti00k_x(x,Rv)

# extinction,Âç≥Ê∂àÂÖâÂáΩÊï∞ A(lambda)
# ‰ª§A_v=1ÔºåÂç≥ÂèØÁîªÂá∫extinction curve
def calzetti00_extinction(wave,A_v,Rv,unit):
	k = calzetti00k_wave(wave,Rv,unit)
	A_lam = k * A_v / Rv
	return A_lam

# Ê±ÇÂá∫ intrinsic E(B-V)
def calzetti00_EBV(ha_lam,hb_lam,ha_flux,hb_flux,Rv,unit):
	ha_k = calzetti00k_wave(ha_lam,Rv,unit)
	hb_k = calzetti00k_wave(hb_lam,Rv,unit)
	empirical_ratio_ha_hb = 2.86
	EBV = (1/(hb_k-ha_k))*2.5*np.log10(ha_flux/hb_flux/empirical_ratio_ha_hb)
	return EBV

# Ê∂àÂÖâ‰øÆÊ≠£ÂáΩÊï∞ ‰ΩøÁî®ÁöÑÊòØBalmer decrement
def calzetti00_dered(wave,flux,ha_lam,hb_lam,ha_flux,hb_flux,Rv,unit):
	k = calzetti00k_wave(wave,Rv,unit)
	EBV = calzetti00_EBV(ha_lam,hb_lam,ha_flux,hb_flux,Rv,unit)
	trans = 10.**(0.4 * EBV * k)
	return flux * trans
```

```python
############################################
####### Ê∂àÂÖâ‰øÆÊ≠£ extinction correction #####
############################################

# ÂÖÉÁ¥†ÁöÑÊ≥¢ÈïøËÆæÂÆö
# ha_6564    hb_4862    nii_6585    oii_3727    oii_3729    oiii_5008    
ha_lam = 0.6564
hb_lam = 0.4862
nii_lam = 0.6585
oii7_lam = 0.3727
oii9_lam = 0.3729
oiii_lam = 0.5008

# Ëé∑ÂèñÂÖÉÁ¥†ÁöÑflux
ha = spaxel_data[:,5]
hb = spaxel_data[:,6]
nii = spaxel_data[:,7]
oii7 = spaxel_data[:,10]
oii9 = spaxel_data[:,9]
# oii9 = spaxel_data[:,7]*0
oiii = spaxel_data[:,8]

E_BV = spaxel_data[:,11]    # Ëé∑ÂèñE(B-V)
Es_BV = 0.44*E_BV    # ËÆ°ÁÆóEs(B-V)

# Âà§Êñ≠ÊòØÂê¶ ha/hb >= 2.86ÔºåËã•ÊòØÔºåÂàô‰øÆÊ≠£ÔºõËã•‰∏çÊòØÔºåÂàô‰∏çÁÆ°
ha_over_hb = ha/hb    # ËÆ°ÁÆóha/hb
tmp_id = np.where(ha_over_hb>=2.86)    # ÊâæÂá∫ ha/hb>=2.86 ÁöÑÊï∞ÁªÑ‰ΩçÁΩÆ
Rv = 4.05    # ËÆ∫Êñá‰∏≠ÁöÑRvÂèñ4.05
def flux_coeff(tmp_lamda):
    
    k_lamda = 2.659*(-2.156+1.509/tmp_lamda-0.198/tmp_lamda**2+0.011/tmp_lamda**3)+Rv
    # ‰øÆÊ≠£Á≥ªÊï∞
    x2 = 0.4*Es_BV*k_lamda
    flux_coff = np.power(10,x2)
    
    return flux_coff

# ËøõË°åfluxÁöÑ‰øÆÊ≠£ÔºÅÔºÅÔºÅ
nii[tmp_id] = nii[tmp_id]*flux_coeff(nii_lam)[tmp_id]
oii7[tmp_id] = oii7[tmp_id]*flux_coeff(oii7_lam)[tmp_id]
oii9[tmp_id] = oii9[tmp_id]*flux_coeff(oii9_lam)[tmp_id]
oiii[tmp_id] = oiii[tmp_id]*flux_coeff(oiii_lam)[tmp_id]


# ËÆ°ÁÆóÂÖÉÁ¥†ÁöÑflux‰πãÊØîÔºåÂπ∂log
lg_oiii_oii = np.log10(oiii/(oii7+oii9))
lg_nii_oii = np.log10(nii/(oii7+oii9))
############################
####### Ê∂àÂÖâ‰øÆÊ≠£ÁªìÊùü #######
############################



# ÁîªÂõæ
# %xdel y
# %xdel x
y = lg_oiii_oii
x = lg_nii_oii

fig = plt.figure()    # ËÆæÂÆöÂõæÂΩ¢Â§ßÂ∞è
ax1 = fig.add_subplot(111)    # ËÆæÂÆöÂ≠êÂõæÔºåË∞ÉÊï¥‰ΩçÁΩÆ

# hist2dÁîªÂõæ
xbins,ybins,hist,cax = ax1.hist2d(x,y,bins=100,range=[[-1.8,0.8],[-2,0.8]],cmap='jet',norm=mpl.colors.LogNorm()) 
# xbins,ybins,hist,cax = plt.hist2d(x,y,bins=100,range=[[-1.8,0.8],[-2,0.8]],cmap='jet',norm=mpl.colors.LogNorm()) 
# ax1.set_aspect(1)
plt.xlabel('Log [N‚Ö°]/[O‚Ö°]',fontsize=16)    # xËΩ¥Ê†áÁ≠æ
plt.ylabel('Log [O‚Ö¢]/[O‚Ö°]',fontsize=16)    # yËΩ¥Ê†áÁ≠æ
# colorbarË∞ÉÊï¥
cbar = fig.colorbar(cax)
cbar.ax.tick_params(which='both',direction='in')
cbar.set_label("Number of all selected spaxels",fontsize=15,labelpad=8)
plt.show() 
# b = np.argwhere(np.isnan(a[:,8]))
# c = np.unique(a[b,0]).astype(int)
# print(lalala.shape)
# print("max = ",np.max(lalala[:,2]))
# print("min = ",np.min(lalala[:,2]))
# print(lalala[20:40,2])
%xdel xbins
%xdel ybins
%xdel hist
%xdel cax
%xdel fig
%xdel ax1
%xdel cbar
```
